```r
a = read.csv(files[1], header = T, stringsAsFactors = F, check.names = F)
names =a[,1]
a = a[, -1]
a = as.matrix(a)
rownames(a)= names 
mat = matrix(0, nrow(a), length(files))
rownames(mat) = rownames(a)
mat[, 1] = a[, 1]
for(i in 2:length(files)){
  a1 = read.csv(files[i], header = T, stringsAsFactors = F, check.names = F)
    names =a1[,1]
a1 = a1[, -1]
a1 = as.matrix(a1)
rownames(a1)= names
  if(all(rownames(a1) == rownames(mat))){
    mat[, i] = a1[, 1]
  }
  else{
    print(paste("Rownames did not match for file", i, sep = "_"))
  }
}
```

```r
names = gsub("GSM[0-9]+_normalized_", "", files)
names = gsub("_.+$", "", names)
```

```r
nam = gsub(".csv", "", files)
si = data.frame(Group = names, row.names = nam)
```

```r
multi_dataset  = function(dataset){
names = rownames(dataset)
dataset = apply(dataset, 2, as.character)
dataset = apply(dataset, 2, as.numeric)
rownames(dataset) = names
head(dataset)
uni = unique(colnames(dataset))
com = combn(uni, 2, simplify = F)
print(com)
for(j in 1:length(com)){
ix = which(colnames(dataset) == com[[j]][1])
ix2 = which( colnames(dataset) == com[[j]][2])
mat = matrix(NA, nrow(dataset), 6)
for(i in 1:nrow(dataset)){
mat[i, 4] = try(t.test(as.numeric(dataset[i, ix]), as.numeric(dataset[i, ix2]))$p.value)
}
mat[, 5] = p.adjust(mat[, 4], method = "fdr")
mean1 = as.numeric(as.character(apply(dataset[, ix], 1, mean, na.rm = T)))
mean2 = as.numeric(as.character(apply(dataset[, ix2], 1, mean, na.rm = T)))
lfc = log2((mean1+0.00000000001)/(mean2 + 0.00000000001)+0.00000000001)
higher = mean1 > mean2
mat[, 1] = mean1
mat[, 2] = mean2
mat[, 3]= lfc
mat[, 6] = higher
colnames(mat) = c("Mean1", "Mean2", "logFC", "pval", "FDR-pval", paste("higher_in_", com[[j]][1], sep = ""))
rownames(mat) = rownames(dataset)
write.table(mat[order(mat[,4]), ], paste("DE_1",com[[j]][1], com[[j]][2], ".tsv", sep = "_" ), quote = F, sep = "\t")
}}
```
