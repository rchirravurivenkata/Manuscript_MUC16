```r
library(pheatmap)
```

```r
overlap = read.csv("GSE71729_series_matrix.csv", header = T, stringsAsFactors = F, check.names = F, comment.char = "#")
head(overlap)

l = overlap[1:7,]
over = overlap
over[-c(1:7), ]
o = over[-c(1:7), ]
all(colnames(o) == colnames(l))
```

```r
## preprocessing the file to identify stromal subtypes
idx = grep("Primary", colnames(o))
primary = o[, idx]
metadata = l[, idx]
rownames(primary) = o[, 1]
metadata =t(metadata)
idx = grep("stroma", metadata[,7])
metadata[idx, 5] = metadata[idx, 7]

##
names = rownames(primary)
primary = apply(primary, 2, as.numeric)
rownames(primary) = names
```

```r
metadata[, 5]
```

```r
cor(as.numeric(primary["MUC16", ]), as.numeric(primary["CGB7", ]))
```

```r
c = c()
for(i in 1:nrow(primary)){
    c = append(c, cor(as.numeric(primary["MUC16",]), as.numeric(primary[i, ]), method = "spearman"))}
```

```r
rownames(primary)[which(c > 0.5)]
```

```r
sampleinfo = data.frame(group = gsub(".+: ", "", as.character(metadata[,5 ])), row.names = colnames(primary))
genes = c("MUC16", "KRT19", "WNT10A", "WNT7B", "PLAT", "PLAU", "MET", "PPL", "IVL", "SCEL")
pheatmap(log2(primary[genes, ]+1),scale ="row", annotation_col = sampleinfo)
```
